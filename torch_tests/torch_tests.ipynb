```python
torch.zeros(5,5) #also torch.ones
```

```python
A = torch.rand(5,5) # same as np.random.rand(5,5)
B = torch.randn(5,5)
torch.matmul(A, B) # torch.dot only for 1D tensors
```

```python
C = torch.rand(5).reshape((1,5))
C_ = A + C
torch.sum(C_, axis = 0) #again, very similar to np
```

```python
img = torch.randn((256, 256, 3))
scaling = torch.tensor([0.5, 1.5, 0.7]) #Typical np.array
img * scaling
```

```python
# If one does 3 imgs
Imgs = torch.randn((3, 256, 256, 3))
Scaling = torch.tensor([1,3,1,3,2,5,6,7,1]).reshape(3, 1, 1, 3)
Imgs * Scaling #Same idea with the broadcasting
# Very similar to np
```

```python
### Unto the interesting stuff (Autograd!)

x = torch.randn((2,5), requires_grad = True)
y = (x**2 / 2).sum()
y.backward() # compute grad
x.grad #(grad of y wrt x)
```
